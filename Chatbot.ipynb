```python
from openai import OpenAI
# openai.api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key="")
client
```

###Upload File

```python
uploaded_file = client.files.create(
    file=open("/content/Deloitte _ Tilintarkastus, konsultointi, rahoitus, riskienhallinta, veropalvelut.html", 'rb'),
    purpose='assistants',
)
```

```python
uploaded_file
```

###Create Assistant

```python
assistant = client.beta.assistants.create(
    name="Deloitte Assistant",
    instructions="You are an AI Assistant that answers any queries related to Deloitte.",
    tools=[{"type": "file_search"}],   # instead of "retrieval"
    model="gpt-4.1"
)
```

###Create Thread

```python
thread = client.beta.threads.create()
```

```python
thread
```

###Create Message Inside Thread

```python
message = client.beta.threads.messages.create(
    thread_id=thread.id,
    role="user",
    content="can you tell me more about Analytics & Artificial Intelligence service they offer ?"
)
```

```python
client.beta.threads.messages.list(thread_id=thread.id).data
# client.beta.threads.messages.list(thread_id=thread.id).data[0].content[0].text.value
```

###Run Assistent

```python
run = client.beta.threads.runs.create(
  thread_id=thread.id,
  assistant_id=assistant.id
)
# run = client.beta.threads.runs.retrieve(thread_id=thread.id, run_id=run.id)
# run
```

###Retrive Assistant's Response

```python
while True:
    run = client.beta.threads.runs.retrieve(thread_id=thread.id, run_id=run.id)
    if run.status=="completed":
        messages = client.beta.threads.messages.list(thread_id=thread.id)
        latest_message = messages.data[0]
        text = latest_message.content[0].text.value
        print(text)
        break;
```
